# DATA FRAME

```python
import pandas as pd
```

```python
data = {
    "Name": ["Thanh Tuan", "Hoang Sa", "Truong Sa"],
    "Score": [2,23,34]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
df = pd.DataFrame(data, columns=['Score', 'Name'])
df
```

```python
#Create new col 
df = pd.DataFrame(data, columns = ['Score', 'Name', 'age'])
df
```

```python
#Change index
df = pd.DataFrame(data, index = ['One', 'two', 'Three' ],
                  columns = ['Score', 'Name', 'age'])
df
```

```python
#View column name, age
df[['Name','age']]
```

```python
#View index
df.loc[['two']] #Location
```

```python
#Filter 
df[df['Score'] == 23]
```

```python
df = pd.DataFrame(data, index = ['One', 'two', 'Three' ],
                  columns = ['Score', 'Name', 'age'])
value_ages = [12,13,15]
df['age'] = value_ages
df
```

```python
#Check and create col pass, if Score >= 5 => True in col pass
df['pass'] = df['Score'] >= 5
df
```

```python
#Delete col pass
'''
LƯU Ý:
- Nếu dùng drop thì DataFrame vẫn còn data
- Nếu dùng del thì mất hết
'''
print(df)
print('--------------')
print('Use drop')
print(df.drop('pass',axis = 1))
print('--------------')
print('Data when use drop')
print(df) 
print('Use del')
del df['pass']
print('--------------')
print(df)
```

```python
#Transpose score
df.T
```

# Lambda and Function

```python
df = pd.DataFrame(
    np.random.randn(4,3),
    columns = list("ABC"),
    index= ['Thanh Tuan', 'Thanh Tu', 'Anh Thu', 'Huong Le']
)
df
```

```python
#Lambda
min_lambda = lambda x: max(x) - min(x)
```

```python
df.apply(min_lambda, axis = 1)
```

```python
#Function
def m2(x):
    return x**2
```

```python
df.apply(m2)
```

# Sorting and Ranking

```python
import pandas as pd
import numpy as np
```

```python
s = pd.Series(range(5), index= ['a', 'e','b', 'c','d'])
s
```

```python
#Sort index (Series)
s.sort_index()
```

```python
df = pd.DataFrame(
    np.arange(12).reshape(3,4),
    index = ['two', 'three', 'four'],
    columns = ['a', 'c', 'd','e']
)
df
```

```python
#Sort index DataFrame
df.sort_index()
```

```python
#Sort columns
df.sort_index(axis= 1, ascending= False)
```

```python
df.sort_values(by ='e')
```

```python
df.sort_values(by =['e', 'a'])
```
